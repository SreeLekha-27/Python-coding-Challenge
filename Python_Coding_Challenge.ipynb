```python
# Retail Inventory Manager
product_categories = ("Electronics", "Stationery", "Clothing")
products = ["Laptop", "Pen", "T-Shirt"]
new_product = input("Enter a new product to add: ")
products.append(new_product)
print("\nCurrent Products:")
for i, item in enumerate(products):
    print(f"{i}: {item}")
remove_index_input = input("Enter the index of the product to remove: ")

if remove_index_input.isdigit():
    remove_index = int(remove_index_input)
    if 0 <= remove_index < len(products):
        removed_item = products.pop(remove_index)
        print(f"Removed product: {removed_item}")
    else:
        print("Invalid index. No product removed.")
else:
    print("Invalid input. Please enter a number.")
print("\n--- Final Inventory ---")
print("Product Categories (Fixed):", product_categories)
print("Final Product List:", products)

```

```python
# BMI Calculator
weight = int(input("Enter your weight in kg: "))
height = float(input("Enter your height in meters: "))
bmi = weight / (height ** 2)
print(round(bmi, 2))
if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal weight")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

```python
# Shopping Cart System

item_count = 0
subtotal = 0

while True:
    price = float(input("Enter item price (Enter 0 to stop): "))
    if price == 0:
        break
    if price < 0:
        print("Invalid price. Please enter a positive value.")
        continue
    item_count += 1
    subtotal += price

else:
    print("Finished adding items")
if subtotal >= 1000:
    discount_rate = 0.20
elif subtotal >= 500:
    discount_rate = 0.10
elif subtotal >= 200:
    discount_rate = 0.05
else:
    discount_rate = 0.00

discount_amount = subtotal * discount_rate
final_total = subtotal - discount_amount

print("\n------ BILL SUMMARY ------")
print("Total Items:", item_count)
print("Subtotal:", subtotal)
print("Discount Applied:", discount_rate * 100, "%")
print("Discount Amount:", discount_amount)
print("Final Total:", final_total)
```

```python
#  Simple Calculator App

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Division by zero is not allowed."
    return a / b

num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))

print("Choose an operation: +, -, *, /, pow, abs, round")
operation = input("Enter the operation: ")

if operation == "+":
    result = add(num1, num2)
elif operation == "-":
    result = subtract(num1, num2)
elif operation == "*":
    result = multiply(num1, num2)
elif operation == "/":
    result = divide(num1, num2)
elif operation.lower() == "pow":
    result = pow(num1, num2)
elif operation.lower() == "abs":
    result = abs(num1)
elif operation.lower() == "round":
    result = round(num1)
else:
    result = "Invalid operation selected."
print("Result:", result)
```

```python
# Global variable to keep track of total transactions
total_transactions = 0

class BankAccount:
    def __init__(self, account_holder, balance=0):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        global total_transactions
        if amount > 0:
            self.balance += amount
            total_transactions += 1
            print(f"Deposited ${amount} to {self.account_holder}'s account.")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount):
        global total_transactions
        if amount <= 0:
            print("Withdrawal amount must be positive!")
        elif amount > self.balance:
            print("Insufficient funds! Withdrawal canceled.")
        else:
            self.balance -= amount
            total_transactions += 1
            print(f"Withdrew ${amount} from {self.account_holder}'s account.")

    def display_balance(self):
        print(f"{self.account_holder}'s current balance: ${self.balance}")


account1 = BankAccount("Alice", 500)

account2 = BankAccount(account_holder="Bob", balance=300)

account1.deposit(200)
account2.deposit(amount=150)

account1.withdraw(100)
account2.withdraw(amount=500)

account1.display_balance()
account2.display_balance()

print(f"Total transactions across all accounts: {total_transactions}")
```
